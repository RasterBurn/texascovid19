```python
import pandas as pd
import numpy as np
from datetime import date, timedelta
from texascovid19 import constants
import os
```

```python
START_DATE=date(2020,3,25)
```

```python
dt = START_DATE
filename = os.path.join(constants.DATA_PATH, f"{dt.isoformat()}.csv")
accum = None

while os.path.exists(filename):
    df = pd.read_csv(filename, index_col='County of Residence', usecols=['County of Residence','Deaths'])
    df.rename(columns={"Deaths": dt.isoformat()}, inplace=True)
    if accum is None:
        accum = df
    else:
        accum = pd.concat([accum, df],axis=1)
    dt += timedelta(days=1)
    filename = os.path.join(constants.DATA_PATH, f"{dt.isoformat()}.csv")
```

```python
accum = accum.fillna(0).astype(np.int32)
accum.index.name = 'County of Residence'
```

```python
accum
```

```python
accum.to_csv(os.path.join(constants.DATA_PATH, f"deaths_timeseries.csv"), header=True, index=True)
```
